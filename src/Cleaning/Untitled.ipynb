```python
ignored_cols = ['product-subtitle', "LOT"]
```

```python
def plotting_features(data, ignored_cols):
    """Return the distribution of numerical features and count of categorical features, specific features will be ignored."""
    cols = [col for col in data.columns if col not in ignored_cols]
    nrows = int(np.ceil(len(cols)/2))
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(30,20), constrained_layout=True)
    ax = ax.ravel()
    for i in range(len(cols)):
        if (data[cols[i]].dtypes == 'object') | (len(data[cols[i]].unique().tolist()) < 10): 

           sns.countplot(y = data[cols[i]], ax=ax[i])
           ax[i].set_title(f'{cols[i]} count')
        else:
           sns.histplot(x = data[cols[i]], ax=ax[i])
           ax[i].set_title(f'{cols[i]} distribution')  
```

```python
plotting_features(rolex, ignored_cols=ignored_cols)
```
